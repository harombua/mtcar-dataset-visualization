## a) Domain Analysis  for mtcars dataset.
### Domain Analysis:-Understanding the features

* Model:- Name of the car.
* mpg:- Miles/ gallon (US)-How many miles it will go per gallon.
* cyl:- No. of cylinders. It’s the power unit of an engine. It’s the chamber where the gasoline is burned and turned into power,Its a strenght of engine.
* Disp:- Displacement.Engine displacement is the measure of the cylinder volume swept by all of the pistons of a piston engine, excluding the combustion chambers.
* Hp:-Horsepower ( hp). It is a unit of measurement of power, or the rate at which work is done, usually in reference to the output of engines or motors. It indicates the power produced by a car's engine.
* drat:-drat: rear axle ratio. Number of turns of drive shaft for every one rotation of wheel axle. Axle ratio refers to the gears in a truck's differential, which connects the rear axle to the driveshaft and then the engine. This ratio expresses a truck's towing capacity and fuel economy, and the relationship between the two. In layman terms it's a rotate ratio e.g. when turning a wide turn or circle how much  car's excel rotates.
* wt:-Weight. Vehicle weight is a measurement of wheeled motor vehicles; either an actual measured weight of the vehicle under defined conditions or a gross weight rating for its weight carrying capacity.
* qsec:-time requried to complete start to the end of a straight 1⁄4 mile.
* vs:-shape of engine
* am:-Transmission
* gear:- Number of forward gears.It is transmission of power.The gears determine the amount of power available from the engine. First gear provides the most pulling power but the least potential for speed, whilst fifth gear which provides the least pulling power allows the greatest range of speed.it transmits engine power and torque can be used by your car's wheels to create motion.
* carb:- Number of venturi in a carburetors.it transmits engine power and torque can be used by your car's wheels to create motion

## Bivariate Analysis

```python
## Scatter plot to compare mpg and hp.

sns.scatterplot(x=data['mpg'],y=data['hp'],s=100,hue=data.cyl, palette=['red','green','orange'])
plt.title('Mileage per Gallon vs Horsepower');
```

### Insights
From the scatterplot above, we see a negative correlation between mpg and hp. It is clear that as the mileage per gallon increases, the horsepower decreases for an engine with a given number of cylinders. This means one has to choose, what the 'gallon of fuel' will do, either make it cover more distance with less power or get more power with less distance.

It is also observed that increase in the number of engine cylinders increase the number of power output(horsepower). As observed from the scatterplot above, the 4-cylinder engines have the lowest horsepower (below 130) with high mpg(upto about 34) records while 8-cylinder engines which were the majority have a high power output(horsepower of above 150) but lowest mpg(below 20). 6-cylinder engines have the compromise of the two variables. 

This means that depending on the tast to be accomplished with the car to be bought/selected, it is good to consider the number of cylinders in the engine of that car. 

```python
## Scatter plot to compare carb and hp.
#palette=['green','orange','brown','dodgerblue','red','black']
sns.barplot(x=data.carb,y=data.hp,hue=data.cyl,palette=['red','blue','black'])
plt.title('Carburetor Barrels (carb) vs Horsepower(hp)');
```

### Insights
From the bar plot above, we can see that there is a positive correlation between the horsepower and the number of barrels in a carburetor. For any given engine (with a given number of cylinders), the more the number of barrels in a carburetor, the higher the horsepower.

From the given data set, it appears that the 4-cylinder engine had ony 1 and 2 barrels in their carburetor, nowonder we have seen that they have less power as compared to 6 and 8-cylinder engines. The 8-cyinder engines have above 2, 3, 4, and 8 barrels per carburetor hence more power. The 6-cylinder is a compromise with 1, 4, and 6 barrels in their carburetors.
